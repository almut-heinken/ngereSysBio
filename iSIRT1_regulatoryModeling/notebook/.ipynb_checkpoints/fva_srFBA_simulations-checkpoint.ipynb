```python
import numpy as np

# Replace NaN values with 0
correlation_results = correlation_results.fillna(0)

# Compute real_correlation based on conditions
def calculate_real_correlation(row):
    mins_corr = row["mins_correlation"]
    maxs_corr = row["maxs_correlation"]

    if mins_corr < 0 and maxs_corr <= 0:
        return max(abs(mins_corr), abs(maxs_corr))  # Case 1
    elif np.sign(mins_corr) != np.sign(maxs_corr): 
        return maxs_corr  # Case 2
    else:
        return maxs_corr  # Case 3 (default)

# Apply function to each row
correlation_results["real_correlation"] = correlation_results.apply(calculate_real_correlation, axis=1)

# Display final DataFrame
print(correlation_results)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

# Collect correlation data
correlation_data = []
for rxn in list(result_df_reactions.index):
    regression = result_df_reactions.loc[rxn].to_dict()
    keys = [float(key.split("_")[-1]) for key in regression.keys()]
    values = list(regression.values())

    # Compute Spearman correlation
    correlation, p_value = spearmanr(keys, values)
    if not np.isnan(correlation):
        correlation_data.append((rxn, correlation, p_value))

# Convert to DataFrame
correlation_df = pd.DataFrame(correlation_data, columns=['Reaction', 'Spearman', 'P-value'])

# Compute -log10(p-value) for better visualization, handle p = 0 case
correlation_df['-log10(P-value)'] = -np.log10(correlation_df['P-value'].replace(0, 1e-15))

# Add jitter to Spearman correlation values
jitter_strength = 0.02
correlation_df['Jittered Spearman'] = correlation_df['Spearman'] + np.random.uniform(
    -jitter_strength, jitter_strength, size=len(correlation_df)
)

# Define thresholds
strong_correlation = 0.5  # Moderate correlation threshold
significance_threshold = 0.01

# Create Volcano Plot
plt.figure(figsize=(12, 8))

# Plot all points with jittered values
plt.scatter(correlation_df['Jittered Spearman'], 
            correlation_df['-log10(P-value)'],
            c='lightgray', alpha=0.3, s=10, label='All Reactions (Non-significant)')

# Highlight significant points
significant = correlation_df['P-value'] < significance_threshold
plt.scatter(correlation_df[significant]['Jittered Spearman'], 
            correlation_df[significant]['-log10(P-value)'],
            c='red', alpha=0.8, s=15, label='Significant Reactions (p < 0.01)')

# Highlight top positive and negative correlations
top_pos = correlation_df.nlargest(5, 'Spearman')
top_neg = correlation_df.nsmallest(5, 'Spearman')

plt.scatter(top_pos['Jittered Spearman'], top_pos['-log10(P-value)'], c='blue', label='Top Positive Correlations', edgecolors='black', s=30)
plt.scatter(top_neg['Jittered Spearman'], top_neg['-log10(P-value)'], c='green', label='Top Negative Correlations', edgecolors='black', s=30)

# Label top reactions
for i, row in pd.concat([top_pos, top_neg]).iterrows():
    plt.text(row['Jittered Spearman'], row['-log10(P-value)'], row['Reaction'], fontsize=9, ha='right')

# Add threshold lines
plt.axhline(-np.log10(significance_threshold), linestyle='--', color='black', label='p = 0.01')
plt.axvline(-strong_correlation, linestyle='--', color='blue', alpha=0.7, label=f'Correlation = {strong_correlation}')
plt.axvline(strong_correlation, linestyle='--', color='blue', alpha=0.7)

# Formatting
plt.xlabel("Spearman Correlation (with jitter)")
plt.ylabel("-log10(P-value)")
plt.title("Volcano Plot of Spearman Correlation for SIRT1 (Jittered)")
plt.legend(loc='upper center')
plt.grid(alpha=0.3)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Define bins for absolute Spearman correlation
bins = [0, 0.5, 0.75, 0.99, 1.0]
labels = ["050%", "5075%", "7599%", "100%"]

# Step 2: Create a column for correlation bins
correlation_df['Correlation Bin'] = pd.cut(correlation_df['Spearman'].abs(), bins=bins, labels=labels, include_lowest=True)

# Step 3: Filter out "exchange" and "transport" subsystems
filtered_df = correlation_df[~correlation_df['Subsystem'].str.contains("exchange|transport", case=False, na=False)]

# Step 4: Loop through each correlation bin and create pie charts
for bin_label in labels:
    # Filter the data for the current bin
    bin_df = filtered_df[filtered_df['Correlation Bin'] == bin_label]

    # Group by subsystems and count the number of reactions
    subsystem_counts = bin_df['Subsystem'].value_counts()

    # Focus on the top 10 subsystems, grouping the rest as "Other"
    top_n = 10
    if len(subsystem_counts) > top_n:
        subsystem_counts = subsystem_counts.nlargest(top_n).append(
            pd.Series({"Other": subsystem_counts.iloc[top_n:].sum()})
        )

    # Create a pie chart
    plt.figure(figsize=(8, 6))
    subsystem_counts.plot(
        kind='pie',
        autopct='%1.1f%%',
        startangle=140,
        colors=plt.cm.tab20.colors,
        wedgeprops={'edgecolor': 'black'}
    )

    # Add a title
    plt.title(f"Subsystem Distribution for {bin_label} Correlation Bin")
    plt.ylabel("")  # Remove the default y-axis label

    # Show the pie chart
    plt.tight_layout()
    plt.show()
```
