```python
# This is a code cell. Adjust and customize as needed.
from collections import deque
graph = {}
for index, row in tft_df.iterrows():
    source = row['source_genesymbol']
    target = row['target_genesymbol']
    if source not in graph:
        graph[source] = []
    graph[source].append(target)

queue = deque(['SIRT1'])
visited = set()

while queue:
    current = queue.popleft()
    if current not in visited:
        visited.add(current)
        neighbors = graph.get(current, [])
        queue.extend(neighbors)

visited_molecules = list(visited)

print("Molecules that SIRT1 interacts with and their downstream interactions:")
print(visited_molecules)
```

```python
# This is a code cell. Adjust and customize as needed.
visited = set()

def traverse_molecules(molecule):
    if molecule not in visited:
        visited.add(molecule)
        neighbors = graph.get(molecule, [])
        for neighbor in neighbors:
            traverse_molecules(neighbor)

traverse_molecules('SIRT1')

sub_df = tft_df[tft_df['source_genesymbol'].isin(visited)]

print(sub_df)
```

```python
# This is a code cell. Adjust and customize as needed.
def create_rule(row):
    if row['consensus_stimulation'] == 1:
        return "("+row['source_genesymbol']+")"
    return "(NOT " + row['source_genesymbol'] + ")"

trn_df['Rule'] = trn_df.apply(lambda row: create_rule(row), axis=1)
```

```python
trn_df
```

```python
# This is a code cell. Adjust and customize as needed.
rule_dict = {}

for index, row in trn_df.iterrows():
    target_gene = row["target_genesymbol"]
    rule = row["Rule"]
    
    if target_gene in rule_dict:
        if rule not in rule_dict[target_gene]:
            rule_dict[target_gene].append(rule)
    else:
        rule_dict[target_gene] = [rule]

result_data = {"Gene": [], "Rule": []}

for target_gene, rules in rule_dict.items():
    if any("(NOT" in rule for rule in rules):
        concatenated_rules = " AND ".join(rules)
    else:
        concatenated_rules = " OR ".join(rules)
    
    result_data["Gene"].append(target_gene)
    result_data["Rule"].append(concatenated_rules)
```

```python
# This is a code cell. Adjust and customize as needed.
result_df = pd.DataFrame(result_data)
```
